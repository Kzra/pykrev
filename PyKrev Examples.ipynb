```python
#Load the BRITE Biological molecules dataframe 
brite_df = pd.read_excel('Brite_DF.xlsx',index_col = 0)
brite_formula = brite_df['F'].to_list()
brite_formula = [x for x in brite_formula if str(x) != 'nan']

brite_counts = pk.element_counts(brite_formula)
brite_ratios = pk.element_ratios(brite_counts)


#Some arbitrary formula groupings to test the plotting functions 
formula_1 = brite_formula[0:300]
formula_2 = brite_formula[200:400]
formula_3 = brite_formula[400:500]

ratios_1 = brite_ratios[0:300]
ratios_2 = brite_ratios[200:400]
ratios_3 = brite_ratios[400:500]

intensities_1 = np.random.rand(len(brite_formula))
```

```python
#Formula Examples
formula = pk.standardize_formula('14N','N',brite_formula)
print('Standardized formula: ',brite_formula,'\n')

formula_counts = pk.element_counts(formula_1)
print('Formula counts: ',formula_counts,'\n')

formula_ratios = pk.element_ratios(formula_counts)
print('Formula ratios: ',formula_ratios,'\n')

dbe = pk.double_bond_equivalent(formula_counts)
print('Double bond equivalence: ',dbe,'\n')

nosc = pk.nominal_oxidation_state(formula_counts)
print('Nominal oxidation state of C: ',nosc,'\n')

ai = pk.aromaticity_index(formula_counts)
print('Aromaticity index: ',ai,'\n')

unique_formula = pk.unique_formula(formula_1,formula_2,formula_3)
print('Unique formula: ',unique_formula,'\n')

missing_formula = pk.missing_formula(formula_1,formula_2,formula_3)
print('Missing formula: ',missing_formula,'\n')
```

```python
#Diversity Examples
ri = pk.relative_intensity(intensities_1)
print('Relative intensities: ',ri, '\n')

print('Diversity indices:  ','\n')
di = pk.diversity_indices(brite_formula,ri)
```

```python
#Plotting Examples

print('Van Krevelen plot (1)', 'Multi Van Krevelen plot (2)', 'Unique plot (3)', 'Missing plot (4)','Missing unique sub plots (5)', 'Van Krevelen Histogram (6)')

fig,ax,cbar,lgd = pk.van_krevelen_plot(brite_ratios,colour='density',ref_reactions=[])
cbar.set_label('Kernel Density')

fig2, ax2, lgd2 = pk.multi_van_krevelen_plot(ratio_lists = [ratios_1,ratios_2,ratios_3],group_labels=['1','2','3'])

fig3, ax3, lgd3, unique_groups_ = pk.unique_plot(formula_1,formula_2,formula_3, group_labels=['1','2','3'])

fig4, ax4, lgd4, missing_groups_ = pk.missing_plot(formula_1,formula_2,formula_3, group_labels=['1','2','3'])

figs, axs, missing_groups, unique_groups = pk.unique_missing_subplots(formula_1,formula_2,formula_3,titles = ['1','2','3'])

histogram, d_index = pk.van_krevelen_histogram(brite_ratios)
```

```python
#Specific analysis of compound classes in the brite dataset 

Organic_Acids = brite_df[brite_df['A'] == 'Organic acids']['F'].dropna()
OA_ratios = pk.element_ratios(pk.element_counts(Organic_Acids))

Lipids = brite_df[brite_df['A'] == 'Lipids']['F'].dropna()
Lipid_ratios = pk.element_ratios(pk.element_counts(Lipids))

Carbohydrates = brite_df[brite_df['A'] == 'Carbohydrates']['F'].dropna()
Carbohydrate_ratios = pk.element_ratios(pk.element_counts(Carbohydrates))

Nucleic_acids = brite_df[brite_df['A'] == 'Nucleic acids']['F'].dropna()
NA_ratios = pk.element_ratios(pk.element_counts(Nucleic_acids))

Peptides = brite_df[brite_df['A'] == 'Peptides']['F'].dropna()
Peptide_ratios = pk.element_ratios(pk.element_counts(Peptides))

Vitamins_and_cofactors = brite_df[brite_df['A'] == 'Vitamins and Cofactors']['F'].dropna()
VC_ratios = pk.element_ratios(pk.element_counts(Vitamins_and_cofactors))

Steroids = brite_df[brite_df['A'] == 'Steroids']['F'].dropna()
S_ratios = pk.element_ratios(pk.element_counts(Steroids))

Hormones_and_transmitters = brite_df[brite_df['A'] == 'Hormones and transmitters']['F'].dropna()
HT_ratios = pk.element_ratios(pk.element_counts(Hormones_and_transmitters))

Antibiotics = brite_df[brite_df['A'] == 'Antibiotics']['F'].dropna()
Antibiotic_ratios = pk.element_ratios(pk.element_counts(Antibiotics))
```

```python
#Multi van krevelen of the Brite dataset - can we draw reference compoynd class polygons from these groups? 

import seaborn as sns
palette = sns.color_palette(None, 10)

fig5, ax5, lgd5 = pk.multi_van_krevelen_plot(ratio_lists = [OA_ratios,Lipid_ratios,Carbohydrate_ratios,NA_ratios,Peptide_ratios,VC_ratios,S_ratios,HT_ratios,Antibiotic_ratios],
                                            colours = palette,
                                            transparency = [0.8] * 9,
                                            symbols = ['o','^','X','D','s','o','^','X','D'], 
                                            edge_colours = [None] * 9, 
                                            group_labels= ['Organic Acids', 'Lipids', 'Carbohydrates','Nucleic acids','Peptides','Vitamins and cofactors','Steroids','Hormones and transmitters', 'Antibiotics'])

fig5.set_size_inches(10,10)
```
